# Support Vector Machines

used when classification into a binary system is needed (2 classes only) as opposed to knn which can classify data into multiple different classes 

in this example of breast cancer, tumors are classified into either malignant 0, or benign 1

a hyperplane (line) is produced to divide the data into 2 categories 
an infinite amount of hyperplanes can be drawn, as such the best hyperplane is chosen using this algorithm (based on the furthest data points from the line)

SVMs are more suitable for datasets with high number of dimensions (features)

```python
import sklearn
from sklearn import datasets
from sklearn import svm 
from sklearn import metrics
import pandas as pd

cancer = datasets.load_breast_cancer(as_frame=True)
x = cancer.data
y = cancer.target 

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.1)

classes = cancer.target_names 
```

```python
import matplotlib.pyplot as plt 

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(x["mean texture"], x["mean radius"], x["mean perimeter"])
```

```python
# interactive map
import plotly.express as px 

fig = px.scatter_3d(x_train, x="mean texture", y="mean radius", z="mean perimeter")
fig.show()
```

```python
classes = ["malignant", "benign"]

clf = svm.SVC(kernel="linear")
clf.fit(x_train, y_train)

y_prediction = clf.predict(x_test)
acc = metrics.accuracy_score(y_test, y_prediction)
print(acc)
```
